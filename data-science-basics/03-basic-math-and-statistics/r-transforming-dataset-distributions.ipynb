# Transforming Dataset Distributions

### Why It's important to Scale Your Data 

* So that different magnitude among variables do not produce erroneous or misleading statistics
* To prepare your data for achine mlearning 

### Two Ways to Scale Your Data

1. **Normalization**: putting each observation on a relative scale between the values of 1 and 0

             Value of observation
    -----------------------------------
    Sum of all observations in variable

2. **Standardization**: rescaling data so that it has a zero mean and unit variance

### Normalizing and Transforming Features With MinMaxScalar() and fit_transform()

```python
mtcars_source = '../../inputs/mtcars.csv'
cars = pd.read_csv(mtcars_source)
```

```python
mpg = cars.mpg 
plt.plot(mpg)
```

```python
mpg.describe()
```

```python
mpg_matrix = mpg.values.reshape(-1,1) # one column matrix

scaled = preprocessing.MinMaxScaler()

scaled_mpg = scaled.fit_transform(mpg_matrix)

plt.plot(scaled_mpg)
# y is now not from 10 to 35 but from 0 to 1
```

```python
scaled_2 = preprocessing.MinMaxScaler(feature_range=(0,10))

scaled_mpg_2 = scaled.fit_transform(mpg_matrix)

plt.plot(scaled_mpg_2)
# y is now not from 10 to 35 but from 0 to 10
```
