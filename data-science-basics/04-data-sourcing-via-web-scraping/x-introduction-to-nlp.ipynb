# Natural Language Processing (NLP)

### Sentence Tokenization

Sentence Tokenization is a process of breaking down paragraphs or a complete set of text into sentences.

### Word Tokenization

Word Tokenization is a process of breaking down a paragraph, a sentence or a complete text corpus into an array of words.

### Stop Words

Stop words are extremly common (thus, general) words that add little to little meaning to a body of text at large.  
Stop words can be thought of as a noise in the corpus of the body of the text.   
F.e. words "a", "these", "they'd", "do", "have", "you"

Advantages of removing stop words:

- Removes noise
- Reduces the size of textual data to be analyzed

### Stemming

Stemming is a process for reducing size of the corpus by converting words to their root word.   
Basically a process for linguistic normalization.   

### Stems

Stems (or word stems) are simply part of the words.

### Example of Stemming

- Raw text: "run", "ran", "running"
- Root word: "run"

### Root Words vs. Base Words

**Root word**: a stem word that may, or may not have meaning when taken alone.
- A root word from which other words are derived.
- The primary form of a word, upon which derivates are made.

**Base word**: a stem word that has meaning, even when taken alone.
- It is a base word that is used outright within a language, with only prefixes or suffixes being attached.
- It is the simpliest form of a word.

### Lemmatizing

Lemmatizing is the process of reducing a word to its base form.   

**Stemming**: 
- raw: "better"
- root: "better"

**Lemmatizing**:
- raw: "better"
- root: "good"

### Part of Speech Tagging

The process of identifying a part of speech within a given body of text.   

### An Example of Part of Speech Tagging

Plain sentence: "Everything is all about resources."

Tagged output: [('Everything','NN'),('is','VBZ'),('all','DT'),('about','IN'),('resources','NNS'),('.','.')]

- **NN** = Noun, singular  
- **VBZ** = Verb, third-person singular present takes   
- **DT** = Determiner   
- **IN** = Preposition/subordinating conjunction   
- **NNS** = Noun plural

### Frequency Distribution Plotting 

Is the initial step towards quantifying a textual corpus.   
It counts and plots the frequency of words in a given text.   


```python
import nltk
```

```python
text = "On Wednesday, the Association for Computing Machinery, the world’s largest society of computing professionals, announced that Hinton, LeCun and Bengio had won this year’s Turing Award for their work on neural networks. The Turing Award, which was introduced in 1966, is often called the Nobel Prize of computing, and it includes a $1 million prize, which the three scientists will share."
```

```python
nltk.download('punkt')
```

 ## Sentence Tokenizer

```python
from nltk.tokenize import sent_tokenize
sent_tk = sent_tokenize(text)
print(sent_tk)
```

## Word Tokenizer

```python
from nltk.tokenize import word_tokenize
word_tk = word_tokenize(text)
print(word_tk)
```

## Removing Stop Words

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
sw = set(stopwords.words('english'))
print(sw)
```

```python
filtered_words = [w for w in word_tk if not w in sw]
print(filtered_words)
```

## Stemming

```python
from nltk.stem import PorterStemmer

port_stem = PorterStemmer()
```

```python
stemmed_words = []

for word in filtered_words:
    stemmed_words.append(port_stem.stem(word))

print(f'Filtered words: \n{filtered_words}\n')
print(f'Stemmed words: \n{stemmed_words}\n')
```

## Lemmatizing

```python
nltk.download('wordnet')
```

```python
from nltk.stem.wordnet import WordNetLemmatizer

lem = WordNetLemmatizer()
stem = PorterStemmer()

lemm_words = []

for word in filtered_words:
    lemm_words.append(lem.lemmatize(word))

print(f'Filtered words: \n{filtered_words}\n')
print(f'Lemmatized words: \n{lemm_words}\n')
```

## Parts of Speech Tagging

```python
nltk.download('averaged_perceptron_tagger')
```

```python
from nltk import pos_tag

pos_tagged_words = pos_tag(word_tk)
print(pos_tagged_words)
```

## Frequency Distribution Plotting

```python
from nltk.probability import FreqDist

fd = FreqDist(word_tk)
print(fd)
```

```python
import matplotlib.pyplot as plt 

fd.plot(30, cumulative=False)
plt.show()
```

```python
fd_cleared = FreqDist(filtered_words)

fd_cleared.plot(30, cumulative=False)
plt.show()
```

```python
fd_alpha = FreqDist(text)
print(fd_alpha)

fd_alpha.plot(30, cumulative=False)
plt.show()
```
