#### Load libraries

```
from pyspark.sql import SparkSession, Row
from pyspark.sql.types import IntegerType, DateType, StringType, StructType, StructField, ArrayType, MapType, DoubleType
from pyspark.sql.functions import lit, col, expr, when, sum, avg, max, min, mean, count, udf, explode, concat_ws
```

#### Create Spark session

```
spark = SparkSession.builder.appName('PySpark Replace NULL/None Values').getOrCreate()
```

#### Create Dataframe

```
data = [
  (1, 704, 'STANDARD' , None, 'PR', 30100),
  (2, 704, None, 'PASEO COSTA DEL SUR' ,'PR', None),
  (3, 709, None, 'BDA SAN LUIS', 'PR', 3700),
  (4, 76166,'UNIQUE', 'CINGULAR WIRELESS', 'TX', 84000),
  (5, 76177,'STANDARD', None, 'TX', None)
]

columns = ['id', 'zipcode', 'type', 'city', 'state', 'population']

df = spark.createDataFrame(data=data, schema=columns)
df.show()
```

#### fillna() & fill() Syntax

```
#Replace 0 for null for all integer columns
df.na.fill(value=0).show()
```

```
#Replace Replace 0 for null on only population column 
df.na.fill(value=0,subset=['population']).show()
```

#### Replace Null/None Value with Empty String

```
df.na.fill('').show()
```

```
df.na.fill('unknown',['city']).na.fill('',['type']).show()
```
