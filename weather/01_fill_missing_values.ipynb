```python
# another way is interpolate
# use neighbour numbers to compute a plausible number for the values that are missing
# f.e. here we miss values 2,3,..,6
xdata = np.array([0,1,4,5,7,8], 'd')
ydata = np.array([10,5,2,7,7.5,10], 'd')

pp.plot(xdata, ydata, '--o')
```

```python
xnew = np.linspace(0,8,9)
ynew = np.interp(xnew, xdata, ydata) # linear interpolation

pp.plot(xdata, ydata, '--o', ms=10)
pp.plot(xnew, ynew, 's')
# as you can now missing values are on the plot
```

```python
xnew = np.linspace(0,8,30) # lets take 30 values betwen 0 and 8
ynew = np.interp(xnew, xdata, ydata) 

pp.plot(xdata, ydata, '--o', ms=10)
pp.plot(xnew, ynew, 's')
```

```python
# lets reload
pasadena = getweather.getyear('PASADENA', ['TMIN', 'TMAX'], 2001)
```

```python
good = ~np.isnan(pasadena['TMIN']) # ~ tilde denotes logical notation so we take not nans
x = np.arange(0, 365)

np.interp(x, x[good], pasadena['TMIN'][good])
```

```python
def fillnans(array):
    good = ~np.isnan(array)
    x = np.arange(len(array))

    return np.interp(x, x[good], array[good])
```

```python
pp.plot(fillnans(pasadena['TMIN']))
pp.plot(fillnans(pasadena['TMAX']))
```
