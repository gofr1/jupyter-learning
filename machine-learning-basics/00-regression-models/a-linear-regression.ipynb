# Regression models

## Simple Linear Regression

```python
import numpy as np 
import pandas as pd 
import matplotlib.pylab as plt 
import sklearn

from pylab import rcParams

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import scale
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 10, 8
```

### Generate some random data

```python
rooms = 2 * np.random.randn(100, 1) + 3
rooms[1:10]
```

```python
price = 265 + 6*rooms + abs(np.random.randn(100, 1))
price[1:10]
```

```python
plt.plot(rooms, price, 'r^')
plt.xlabel('Number of rooms average')
plt.ylabel('Average home price, 1000s USD')
plt.show()
```

### Doing simple regression

```python
x = rooms
y = price 

LinReg = LinearRegression()
LinReg.fit(x,y)

print(f"""Simple Algebra

- y = mx + b
- b = intercept = {LinReg.intercept_[0]:.2f}

Estimated Coefficients

- LinRef .coef = [{LinReg.coef_[0][0]:.2f}] Estimated coefficient for the terms in the linear regression problem.""")
```

```python
print(f' The score function returns the coefficient of determination which is the r squared of the prediction.\n In this case, we\'re getting {LinReg.score(x,y):.5f}.\n So our linear regression model is performing really well, our r squared value is close to one and that\s a good thing.')
```
