# Decision tree models with CART

A decision tree is a decision-support tool that models decisions in order to predict probable outcomes of those decisions.

Decision trees are comprised of three types of nodes:
- Root node
- Decision node
- Leaf node

A **root node** is node that represents the entire population or sample (this gets divided into two or more homogenous sets).  
A **decision node** is where a subnode splits into further subnodes.  
And a **leaf node** is just the end node or the furthest subnode.

## Decision tree algorithms

These are a class of supervised machine learning methods that are useful for making predictions from nonlinear data.

Decision tree algorithms are an appropriate fit for:

- *continuous* input and/or output variables
- *categorical* input and output variables

## Types of decision trees

1. **Categorical variable** decision tree (or classification trees): when you use a decision tree to predict for a categorical target variable.
2. **Continuous variable** decision tree (or regression trees): when you use a decision tree to predict for continuous target variable.

## Main benefits of using decision trees.

- Decision trees are nonlinear models
- Decision trees are very easy to interpret
- Trees can be easily represented graphically (in a way that's easy to read and easy to interpret)
- Decision trees require less data preparation

## Assumptions of decision trees algorithms

- Root node = Entire training set
- Predictive features are either categorical, or (if continuous), they have to be binned prior to model deployment
- Rows in the dataset have a recursive distribution based on the values of attributes

## How decision trees algorithms work

1. Decision trees deploy recursive binary splitting to stratify or segment the predictor space into a number of simple, non-overlapping regions
2. Make a prediction for a given observation by using the mean or mode of the training observation in the region to which it belongs
3. Use a set of splitting rules to summarize the tree

## Recursive binary splitting

Is the process that's used to segment a predictor space into regions in order to create a binary decision tree.  
Recursive binary splitting stops when user-defined criteria are met.  

### How recursive binary splitting works

At every stage we split the region into two, and we do this by the following criteria:

- In regression trees: use the sum square of errors (SSE) to calculate the loss function, thus identifying the very best split.
- In classification trees: use the Gini index to calculate the loss function, thus identifying the best split.

### Characteristics of recursive binary splitting

- Top-down (it begins at the top of a tree, at which point all observations belong to a single region, and then successfully splits the predictor space. Each split is indicated via two new branches further down on the tree.)
- Gready (because at each step of the tree-building process, the best split is made at that particular step rather than looking ahead and picking a split that will lead to a better tree in some future step.)

## Disadvantages of decision tree have algorithm

- Non-robust
- Sensitive to training data.
- Globally optimum tree is not guaranteed

## Using decision trees for regression

Regression trees are appropriate when:

- Target is a continuous variable
- Linear relationship between features and target

Output values from terminal nodes represent the mean response:

- Values of new data points will be predicted from that mean

## Recursive splitting in regression trees

At every stage in a regression tree, the region is split into two according to the sums of square errors:

The model begins with the entire data set, S, and searches every distinct value of every predictor to find the predictor and split value that partition the data into two sub groups, S1 and S2, such that the overall sum of square errors are minimized.

## Using decision trees for classification

Classification trees are appropriate:

- Target is a binary, categorical variable

Output values from terminal nodes here represent the mode response:

- Values of new data points will be predicted from that mode

## Recursive splitting in classification trees

At every stage in a classification tree, the region is split into two according to a user-defined metric.

- F.e. the Gini index. This is a measure of total variance across the K classes. It measures the probability of misclassification.

## Tree pruning 

Is the process that's used to overcome model overfitting by removing subnodes of a decision tree. In other words, replacing a whole subtree by a leaf node.

### Why you need to prune a decision tree?

If you grow a tree deep, it would make good predictions on the training data but will probably lead to bad performance on the test data.  
That's because of overfitting. In these cases, you need to prune the tree.

### Two popular tree pruning methods include:

- Hold-out test 
- Cost-complexity pruning

*Hold-out test* is the fastest and simplest pruning method. With this method, you work through each leaf node in the tree and evaluate the effect original removing it using a hold-out test set. A leaf node is removed only if removing it results in a drop in the overall cost function on the entire test set. You stop removing nodes when no further improvements can be made.

*Cost-complexity pruning* is also called the weakest link pruning. This is wear a learning parameter, alpha, is used to weigh weather nodes can be removed based on the size of the subtree.


```python
import pandas as pd
import numpy as np
from sklearn import datasets

from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree

import matplotlib.pyplot as plt
```

```python
np.set_printoptions(precision=4, suppress=True)
%matplotlib inline
plt.rcParams['figure.figsize'] = (15, 13)
plt.style.use('seaborn-whitegrid')
```

```python
iris = datasets.load_iris()
X = pd.DataFrame(iris.data)
y = iris.target
```

```python
model = DecisionTreeClassifier()
model.fit(X,y)
```

```python
plot_tree(model, feature_names=iris.feature_names, filled=True, class_names=iris.target_names);
```
