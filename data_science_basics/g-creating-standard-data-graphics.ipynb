# Creating standard data graphics

## Functional method


```python
import numpy as np 
from numpy.random import randn
import pandas as pd 
from pandas import Series, DataFrame

import matplotlib.pyplot as plt 
from matplotlib import rcParams
```

## Creating a line chart from a list of objects

### Plotting a line chart in matplotlib

```python
x = range(1,10)
y = [1,2,3,4,0,4,3,2,1]

plt.plot(x,y)
```

### Ploting a line chart from a Pandas object

```python
mtcars_source = '../inputs/mtcars.csv'

cars = pd.read_csv(mtcars_source)
mpg = cars['mpg']
```

```python
mpg.plot()
```

```python
df = cars[['cyl', 'wt', 'mpg']]
df.plot()
```

### Creating a bar chart

```python
plt.bar(x, y)
```

### Creating a bar chart from Pandas objects

```python
mpg.plot(kind="bar")
```

```python
mpg.plot(kind="barh")
```

```python
df.plot(kind="bar")
```

### Creating a pie chart

```python
x = [1, 2, 3, 4, .5]
plt.pie(x)
plt.show()
```

### Saving a plot

```python
plt.pie(x)
plt.savefig('pie-chart.png')
```
