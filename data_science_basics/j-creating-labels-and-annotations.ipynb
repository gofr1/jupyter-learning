```python
import numpy as np 
from numpy.random import randn
import pandas as pd 
from pandas import Series, DataFrame

import matplotlib.pyplot as plt 
from matplotlib import rcParams
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 8,4
```

### Labeling plot feature

Functional method

```python
x = range(1,10)
y = [1,2,3,4,.5,4,3,2,1]
plt.bar(x,y)

plt.xlabel('your X-axis label')
plt.ylabel('your Y-axis label')
```

```python
z = [1,2,3,4,.5]
veh_type = ['bicycle', 'motorbike', 'car', 'van', 'stroller']

plt.pie(z, labels=veh_type)
plt.show()
```

The object-oriented method

```python
mtcars_source = '../inputs/mtcars.csv'
cars = pd.read_csv(mtcars_source)

mpg = cars.mpg

fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

mpg.plot()

ax.set_xticks(range(32))
ax.set_xticklabels(cars.model, rotation=60, fontsize='medium')
ax.set_title('Miles per Gallon of Cars in mtcars Dataset')

ax.set_xlabel('model')
ax.set_ylabel('miles/gallon')
```
