# Parametric correlation analysis

A method you can use to find correlation between linearly related continuous numeric variables

** correlation doesn't mean causation **

## Pearson Correlation Coefficient

R = 1 -> Strong positive relation  
R = 0 -> Not linearly correlated  
R = -1 -> Strong negative relation  

It assumes that:

* your data is normally distributed
* you have continuous, numeric variables
* your variables are linearly related

Use it:

* to uncover (linear) relationships between variables
* not to rule out possible (nonlinear) relationships between variables

```python
import numpy as np 
import pandas as pd 

import matplotlib.pyplot as plt 
import seaborn as sb 
from pylab import rcParams

import scipy
from scipy.stats.stats import pearsonr
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 8, 4
plt.style.use('seaborn-whitegrid')
```

### The Pearson Correlation

```python
mtcars_source = '../inputs/mtcars.csv'
cars = pd.read_csv(mtcars_source)
```

```python
sb.pairplot(cars)
```

```python
x = cars[['mpg', 'hp', 'qsec', 'wt']]
sb.pairplot(x)
```

## Using scipy to calculate the Pearson correlation coefficient

```python
mpg = cars.mpg
hp =cars.hp
qsec = cars.qsec
wt = cars.wt

# R = 1 -> Strong positive relation
# R = 0 -> Not linearly correlated
# R = -1 -> Strong negative relation

pearsonr_coefficient, p_value = pearsonr(mpg, hp)
print(f'PearsonR correlation coefficient {pearsonr_coefficient:0.3f}')
```

```python
pearsonr_coefficient, p_value = pearsonr(mpg, qsec)
print(f'PearsonR correlation coefficient {pearsonr_coefficient:0.3f}')
```

```python
pearsonr_coefficient, p_value = pearsonr(mpg, wt)
print(f'PearsonR correlation coefficient {pearsonr_coefficient:0.3f}')
```
