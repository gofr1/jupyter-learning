# Visualizing time series

A time series is a measure of unit change over time for any variable under observation

```python
import numpy as np 
from numpy.random import randn
import pandas as pd 
from pandas import Series, DataFrame

import matplotlib.pyplot as plt 
from matplotlib import rcParams
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 15,4
```

### The simplest time series plot

```python
source = '../inputs/superstore-sales.csv'
df = pd.read_csv(source, index_col='Order Date', parse_dates=True)
df.head()
```

```python
df['Order Quantity'].plot()
# Too much data
```

```python
# Let's create a sample
df_sampled = df.sample(n=100, random_state=25, axis=0)

plt.xlabel('Order Date')
plt.ylabel('Order Quantity')
plt.title('Superstore Sales')

df_sampled['Order Quantity'].plot()
```
